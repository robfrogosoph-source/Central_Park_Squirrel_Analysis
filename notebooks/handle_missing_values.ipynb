# Import Required Libraries
Import necessary libraries such as pandas and numpy for data manipulation.

```python
# Import Required Libraries
import pandas as pd
import numpy as np
```

# Load the Dataset
Load a dataset into a pandas DataFrame, for example, from a CSV file.

```python
# Load the Dataset
df = pd.read_csv('../data:/2018_Central_Park_Squirrel_Census_-_Squirrel_Data_20260422.csv')
df.head()
```

# Identify Missing Values
Use methods like isnull() and sum() to detect and count missing values in the DataFrame.

```python
# Identify Missing Values
missing = df.isnull().sum()
print(missing)
```

# Drop Missing Values
Use dropna() to remove rows or columns with missing values based on specified criteria.

# Fill Missing Values
Use fillna() to impute missing values with methods like mean, median, or forward/backward fill.

```python
# Fill Missing Values
# Fill categorical columns with mode
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)

# For numerical columns, fill with mean
numerical_cols = df.select_dtypes(include=[np.number]).columns
for col in numerical_cols:
    df[col].fillna(df[col].mean(), inplace=True)

print("Filled missing values")
```

# Verify Handling of Missing Values
Check the DataFrame again for missing values and summarize the changes made.

```python
# Verify Handling of Missing Values
missing_after = df.isnull().sum()
print("Missing values after handling:")
print(missing_after)
print("Total missing before:", missing.sum())
print("Total missing after:", missing_after.sum())
```

# Save the Cleaned Dataset

```python
# Save the Cleaned Dataset
df.to_csv('../data:/cleaned_squirrel_data.csv', index=False)
print("Cleaned data saved.")
```
